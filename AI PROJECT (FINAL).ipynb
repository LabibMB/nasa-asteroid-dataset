### PLOTS

```python
#Matplotlib histogram to visualize the asteroid absolute magnitude distribution. 
#This could help to visualize the distribution of asteroid brightness.
plt.figure(figsize=(8,5))
abs_mag = df['Absolute Magnitude']
plt.hist(abs_mag, bins=20, color='green')
plt.xlabel('Absolute Magnitude')
plt.ylabel('Count')
plt.title('Distribution of Asteroid Absolute Magnitudes')
plt.show()
```

```python
#Seaborn scatter plot to visualize the relationship between asteroid size and miss distance. 
#This could help to understand whether larger asteroids tend to pass closer or further from Earth.
plt.figure(figsize=(8,8))
size = df['Est Dia in KM(max)']
miss_dist = df['Miss Dist.(kilometers)']
sns.scatterplot(x=size, y=miss_dist)
plt.xlabel('Maximum Estimated Asteroid Diameter (km)')
plt.ylabel('Miss Distance from Earth (km)')
plt.title('Relationship Between Asteroid Size and Miss Distance')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler( random_state=9)
x_resampled, y_resampled = ros.fit_resample(x, y) #over sampled minority
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_resampled,y_resampled,test_size=0.3,random_state= 9)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit_transform(x_train)
scaler.transform(x_test)
```

Creating a gridsearch model for SVC

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.svm import SVC
svm = SVC()
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001],'kernel': ['rbf']} 
grid = GridSearchCV(svm,param_grid)
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
grid.best_params_
```

```python
y_pred = grid.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()

dt_model = dt_model.fit(x_train,y_train)

```
